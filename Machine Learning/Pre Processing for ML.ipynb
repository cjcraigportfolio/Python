```python
ct1 = pd.crosstab(bank_categorical.gender,bank_categorical.pep,margins = False)
ct2 = pd.crosstab(bank_categorical.region,bank_categorical.pep,margins = False)
ct3 = pd.crosstab(bank_categorical.married,bank_categorical.pep,margins = False)
ct4 = pd.crosstab(bank_categorical.car,bank_categorical.pep,margins = False)
ct5 = pd.crosstab(bank_categorical.savings_acct,bank_categorical.pep,margins = False)
ct6 = pd.crosstab(bank_categorical.current_acct,bank_categorical.pep,margins = False)
ct7 = pd.crosstab(bank_categorical.mortgage,bank_categorical.pep,margins = False)

print(ct1)
print('\n',ct2)
print('\n',ct3)
print('\n',ct4)
print('\n',ct5)
print('\n',ct6)
print('\n',ct7)
```

### 11. Now consider a modified version of the bank data that contains missing values (bank_data_missing_vals.csv). 

### (a) Using Pandas determine all the attributes with missing values and the number of missing values for each such attribute. 

### (b) Show all the instances in the data that contain a missing value. 

### (c) Fill the missing values for all numeric attributes using the mean value for the attribute.

### (d) After filling in the missing numeric values, drop all rows where a categorical attribute contains a missing value. 

### (e) Show that the final resulting table does not contain missing values.

```python
#displaying overview of bank_data_missing_vals
bank_missing = pd.read_csv('bank_data_missing_vals.csv',index_col = 0,na_values = ["?"]) # For ID
bank_missing.head()
```

```python
#a T/F
bank_missing.isnull().any()
```

```python
#b Number
bank_missing.isna().sum()
```

```python
bank_missing.info() #another way to see nulls
```

```python
#b
bank_missing[bank_missing.isna().any(axis = 1)] #all instances
```

```python
#c
#fill age & income with mean
age_mean = bank_missing.age.mean()
bank_missing.age.fillna(age_mean,axis=0,inplace=True)

income_mean = bank_missing.income.mean()
bank_missing.income.fillna(income_mean,axis=0,inplace=True)

bank_missing.head()
```

```python
#d
#drop gender and region na's
bank_missing.drop(bank_missing[bank_missing.gender.isnull()].index,axis=0,inplace = True)
```

```python
bank_missing.drop(bank_missing[bank_missing.region.isnull()].index,axis=0,inplace = True)
```

```python
#e
bank_missing.isna().sum() #no more nulls
```
