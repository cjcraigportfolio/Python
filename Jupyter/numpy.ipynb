```python
import numpy as np
```

```python
# a NumPy array with 100 equally spaced values in the range 0 through 100(not including 100).
a=np.arange(0,100)
print(a)
```

```python
#a NumPy array with 10 equally spaced values in the range 0 through 10
b=np.arange(0,100,10)
print(b)
```

```python
#a NumPy array in the range 0 through 10 (inclusive) with values spaced at 0.
c=np.linspace(0,10,100)
print(c)
```

```python
#a random two-dimensional array with the dimensions 10 b
d=np.random.rand(10,10)
print(d)
```

```python
#reshape a so that it is a two-dimensional array with the dimensions 10 by 10

a=np.reshape(a,(10,10))
print(a)
```

```python
#shows the fourth row and the fifth column, thus getting 45
print(a[4,5])
```

```python
#shows the 4th row
print(a[4])
```

```python
#sum of d
print(np.sum(d))
```

```python
#sum of a
print(np.max(a))
```

```python
#transpose of b
print(np.transpose(b))
```

```python
#adding a and d
print(a+d)
```

```python
#multiplying a*d
print(a*d)
```

```python
#dot product a nd d
print(np.dot(a,d))
```
